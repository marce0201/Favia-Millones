##  <a id='t8'>8. PANDAS (Python Data Analysis)</a>

**¿Qué es un dataset?**

Español: "conjunto de datos"

Un conjunto de datos o dataset corresponde a los contenidos de una única tabla de base de datos o una única matriz de datos de estadística.  
- columna: representa una variable en particular
- fila: representa a un miembro determinado del conjunto de datos que estamos tratando.

*Big Data:* conjuntos de datos tan grandes que las aplicaciones de procesamiento de datos tradicionales no los pueden procesar debido a la gran cantidad de datos contenidos en la tabla o matriz.

Tipos:  estructurado, semi-estructurado y no-estructurado.

**¿Qué es un dataframe?**

Español: “hoja de datos o marco de datos”

Se utiliza normalmente cuando se realiza un estudio estadístico sobre los objetos de una muestra. Cada fila corresponde a un objeto de la muestra y cada columna a una variable.

*Diferencia:* Un dataframe es un dataset que a la vez está organizado en columnas, de modo que en el dataframe tendremos los datos estructurados y cada columna con su nombre correspondiente. Ello facilitará y hará más sencillo consultar, modificar o transformar el conjunto de datos contenidos en la hoja de datos.

Un DataFrame en Pandas es una **estructura de datos tabular bidimensional que se utiliza para almacenar y manipular datos en Python**. Es similar a una hoja de cálculo o una tabla de base de datos, donde los datos se organizan en filas y columnas. Cada columna puede contener datos de diferentes tipos (números, texto, fechas, etc.).

### 8.4. <a id='8.4.'>Concatenaciones y merge</a>

| Método                                | Descripción                                                      |
|---------------------------------------|------------------------------------------------------------------|
| Concatenaciones          | Combina DataFrames apilando filas (vertical) o pegando columnas (horizontal) uno debajo del otro o al lado.                              |
| Merge                    | Combina DataFrames en función de una o más columnas comunes.                             |

**Utilidad**

- Concatenaciones: Útil cuando se cuenta con DataFrames con la misma estructura y se desea unirlos sin considerar valores coincidentes.
- Merge: Útil cuando se desea combinar datos en función de valores coincidentes en columnas específicas.

#### Concatenación vertical

```python
data_3 = {'A': [1, 2, 3], 'B': [4, 5, 6]}
df_3 = pd.DataFrame(data_3)

data_4 = {'A': [7, 8], 'B': [9, 10]}
df_4 = pd.DataFrame(data_4)
```

```python
df_3
```

```python
df_4
```

```python
df_concat_vertical = pd.concat([df_3, df_4], axis=0)
df_concat_vertical
```

#### Concatenación horizontal

```python
data_5 = {'A': [1, 2, 3], 'B': [4, 5, 6]}
df_5 = pd.DataFrame(data_5)

data_6 = {'C': [7, 8, 9], 'D': [10, 11, 12]}
df_6 = pd.DataFrame(data_6)
```

```python
df_5
```

```python
df_6
```

```python
df_concat_horizontal = pd.concat([df_5, df_6], axis=1)
df_concat_horizontal
```

#### Merge

image.png

```python
data_7 = {'clave': ['A', 'B', 'C'], 'valor1': [1, 2, 3]}
df_7 = pd.DataFrame(data_7)

data_8 = {'clave': ['B', 'C', 'D'], 'valor2': [4, 5, 6]}
df_8 = pd.DataFrame(data_8)
```

```python
df_7.to_csv( '/content/sample_data/df_7.csv' )
df_8.to_csv( '/content/sample_data/df_8.csv' )
```

```python
df_7.to_excel( '/content/sample_data/df_7.xlsx' )
df_8.to_excel( '/content/sample_data/df_8.xlsx')
```

```python
# on: nombre de variable usado para hacer merge
# how: mantener variables similares (inner) o mantener todos los valores (outer)

df_merged_1 = pd.merge(df_7, df_8, on='clave', how='inner')
df_merged_1
```

```python
# on: nombre de variable usado para hacer merge
# how: mantener variables similares (inner) o mantener todos los valores (outer)

df_merged_1 = pd.merge(df_7, df_8, on='clave', how='outer')
df_merged_1
```

##  <a id='t9'>9. NUMPY (Numerical Python)</a>

Los datos, presentados en forma de números, se tratan como **arrays (cuadros)** para las funciones multidimensionales y las operaciones de reordenación. Se trata de una herramienta muy utilizada en el campo de la Data Science.

Muchas técnicas de Data Science requieren cuadros y matrices de gran tamaño y cálculos complejos para extraer información valiosa de los datos. NumPy simplifica este proceso con funciones matemáticas diversas y variadas.

#### Arrays de 1 dimension (array 1D ):

image.png

Se trata de una **biblioteca** utilizada para realizar **cálculos numéricos y científicos**. Proporciona un objeto de **matriz multidimensional** llamado `ndarray` que es más eficiente y conveniente para operaciones matemáticas en comparación con las listas de Python estándar.

| Método                   | Descripción                                              |
|--------------------------|------------------------------------------------------------------|
| `np.array(a)`             | Crea un array np de -dimensiones a partir de la secuencia a.                            |
| `np.linspace(a, b, N)`             | Crea un array np 1D con N valores equiespaciados <br> desde a hasta b (inclusivamente). |
| `np.arange(a, b, step)`             | Crea un array np 1D con valores desde a hasta b (exclusivamente) <br> incrementados por step.                       |
| `np.zeros(N)`  | Crea un array np 1D de ceros de longitud N.                           |
| `np.zeros((n, m))`             | Crea un array np 2D de ceros con n filas y m columnas.                            |
| `np.ones(N)`             | Crea un array np 1D de unos de longitud N. |
| `np.ones((n, m))`             | Crea un array np 2D de unos con n filas y m columnas.                       |
| `np.eye(N)`  | Crea un array np 2D con n filas y m columnas <br> con unos en la diagonal (matriz identidad de tamaño N).                           |
| `np.concatenate()`             | Une una secuencia de arrays a lo largo de un eje existente.                            |
| `np.hstack()`             | Apila arrays en secuencia horizontalmente (columna por columna). |
| `np.vstack()`             | Apila arrays en secuencia verticalmente (fila por fila).                       |
| `np.column_stack()`  | Apila arrays 1-D como columnas en un array 2-D.                           |
| `np.random.normal(`             | Genera muestras aleatorias de una distribución normal (Gaussiana). |
| `np.linalg.inv()`             | Calcula el inverso (multiplicativo) de una matriz.                       |
| `np.dot()` o `@`  | Multiplicación de matrices.                           |

###  Accediendo a datos de arrays numpy

Por ejemplo, si X es un array de 2 dimensiones, entonces :
 - X[ i , : ] es la fila i del array.
 - X[ : , j ] es la columna j del array
Cuando se especifica una dimension y todas las otras dimensiones de la derecha son :, entonces es opcional colocar los simbolos : Por ejemplo
  - en un array 2D,  X[ i, : ] es lo mismo que X[ i ]  (fila i)
  - en un array 3D,  X[ i, : , : ] es lo mismo que X[ i ] (matriz i)

```python
miarray = np.array([[11, 12, 13, 14], [21, 22, 23, 24], [31, 32, 33, 34]])
print(miarray)
```
